```python
frame = cv2.imread("C:/Users/satwi/Desktop/Satwik/MyPython/ML Bootcamp/thor.jpg")
```

```python
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(frame)
```

```python
classifier = cv2.CascadeClassifier("C:/Users/satwi/Desktop/Satwik/MyPython/ML Bootcamp/haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces[0]
```

```python
x, y, w, h = face
```

```python
face = frame[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 4)
```

```python
plt.imshow(out)
```
